```python


import sys
import newsapi
apikey = 'c69ca2337f8b44a1ade92e8c0b151280'
from newsapi import NewsApiClient
import numpy as nd
# Init
newsapi = NewsApiClient(api_key='c69ca2337f8b44a1ade92e8c0b151280')

nosofresults = 20
# User Input on the search word and number of items. 

search_term = input("Welome to Balanced View !!!! \nWhat do you want to search about: ")
# the main search function accessing a term
data = newsapi.get_everything(q=search_term,
                              sort_by='relevancy',
                              page_size = nosofresults)

# converting the output of search into an array and adding new columns Sentiment score and variance
import pandas as pd
data = pd.DataFrame.from_dict(data)
data = pd.concat([data.drop(['articles'], axis=1), data['articles'].apply(pd.Series)], axis=1)
data['SentimentScorecontent']= 0.0
data['Variance']=0.0
data['content'] = data['content'].astype(str)


#importing textblob to provide the sentiment score
from textblob import TextBlob
for i in range (1,nosofresults):
    i=1
    while True:
        data.SentimentScorecontent[i]= TextBlob(data.content[i]).sentiment.polarity
        i= i+1
        if i==nosofresults:
            break

# adding Variance 
for i in range (1,nosofresults):
    i=1
    while True:
        data.Variance[i]= (data.SentimentScorecontent[i]-data['SentimentScorecontent'].mean())* (data.SentimentScorecontent[i]-data['SentimentScorecontent'].mean())
        i= i+1
        if i==nosofresults:
            break

data = data.drop(['status','totalResults','urlToImage','author','content'], axis=1)


# sorting using variance
data = data.sort_values(by='Variance', ascending= False)
data.to_html(open('searchoutput.html', 'w'))
data.head(nosofresults)


```
